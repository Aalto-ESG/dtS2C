```python
import os
import pandas as pd

# Specify the folder path where your .feather files are located
folder_path = 'intermediate'

# Get a list of all .feather files that start with "psutil"
def merge_feathers(name_prefix):
    feather_files = [file for file in os.listdir(folder_path) if file.startswith(name_prefix) and file.endswith('.feather')]
    
    # Initialize an empty DataFrame to store the merged data
    merged_df = pd.DataFrame()
    
    # Read and merge each .feather file
    for file in feather_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_feather(file_path)
        merged_df = pd.concat([merged_df, df], ignore_index=True)
    
    # Now `merged_df` contains the merged data from all .feather files
    if "rapl_joules" in merged_df.columns:
        merged_df = merged_df.explode("rapl_joules")
    return merged_df.drop_duplicates(keep='first')

df_psutil = merge_feathers("psutil")
df_carla = merge_feathers("carla")
df_carla
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import os

# Set the style to white background
plt.rcParams.update({
    'font.size': 14,
    # 'font.weight': 'bold',         # Bold legend and ticks
    # 'axes.titleweight': 'bold',    # Bold titles only
    # 'axes.labelweight': 'bold',    # Bold axis labels only
    'axes.titlesize': 18,
    'axes.labelsize': 18,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 14,
    'figure.figsize': (6, 4),

})
sns.set_style("whitegrid")
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.rcParams['axes.spines.left'] = False

x_axis_label = "Number of cameras"

df = df_carla.copy()
df[x_axis_label] = df["num_cameras"]
df["throughput"] = df["num_cameras"] * df["num_frames"] / df["exec_time"]
df["fps"] = df["num_frames"] / df["exec_time"]

# Function to extract numerical value from sensor size string (e.g., "640x640" -> 640)
def extract_size_value(size_str):
    return int(size_str.split('x')[0])

# Sort sensor sizes by their actual numerical values
sensor_sizes = sorted(df["sensor_size"].unique(), key=extract_size_value)
print("Sorted sensor sizes:", sensor_sizes)  # This will show the correct order

# Create custom color palette (blue=smallest, green=largest)
blue = '#1f77b4'
orange = '#ff7f0e'
green = '#2ca02c'
red = '#d62728'
custom_colors = [blue, orange, red, green][:len(sensor_sizes)]  # Adjust based on number of sizes
color_palette = dict(zip(sensor_sizes, custom_colors))

# First plot with custom legend
sns.lineplot(df, hue="sensor_size", x=x_axis_label, y="throughput", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
os.makedirs("plots", exist_ok=True)
# plt.title("Image throughput")
plt.ylabel("Throughput (images/s)")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/local_throughput.png")
plt.savefig("plots/local_throughput.pdf")
plt.show()

# Second plot with custom legend
sns.lineplot(df, hue="sensor_size", x=x_axis_label, y="fps", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("Simulation frames per second")
plt.ylabel("FPS")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/local_fps.png")
plt.savefig("plots/local_fps.pdf")
plt.show()

df["watts"] = df["rapl_joules"] / df["rapl_duration"]
micro_joule_to_joule = 10**-6
df["joule_per_image"] = df["rapl_joules"] / (df["num_frames"] * df["num_cameras"]) * micro_joule_to_joule

sns.lineplot(df, hue="sensor_size", x=x_axis_label, y="watts", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("Average CPU power (watts)")
plt.ylabel("Power (watts)")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/cpu/avg_cpu_power.png")
plt.savefig("plots/cpu/avg_cpu_power.pdf")
plt.show()

sns.lineplot(df, hue="sensor_size", x=x_axis_label, y="rapl_joules", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
plt.show()

sns.lineplot(df, hue="sensor_size", x=x_axis_label, y="joule_per_image", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("CPU energy consumption per image (joules)")
plt.ylabel("Joules per image")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/cpu/cpu_power_per_image.png")
plt.savefig("plots/cpu/cpu_power_per_image.pdf")
plt.show()
```

```python
def timestamp_to_sensors(timestamp):
    for row in df_carla.iterrows():
        num_cameras = row[1]["num_cameras"]
        start = row[1]["start_timestamp"]
        end = row[1]["end_time"]
        if start <= timestamp <= end:
            return num_cameras
    return None

def timestamp_to_sensor_size(timestamp):
    for row in df_carla.iterrows():
        sensor_size = row[1]["sensor_size"]
        start = row[1]["start_timestamp"]
        end = row[1]["end_time"]
        if start <= timestamp <= end:
            return sensor_size
    return None

df_psutil["num_cameras"] = df_psutil["timestamp"].apply(timestamp_to_sensors)
df_psutil[x_axis_label] = df_psutil["num_cameras"]
df_psutil["sensor_size"] = df_psutil["timestamp"].apply(timestamp_to_sensor_size)
df_psutil["GPU"] = df_psutil["GPU Percent"] * 100

# Function to extract numerical value from sensor size string (e.g., "640x640" -> 640)
def extract_size_value(size_str):
    if size_str is None:
        return 0
    return int(size_str.split('x')[0])

# Sort sensor sizes by their actual numerical values
sensor_sizes = sorted(df_psutil["sensor_size"].dropna().unique(), key=extract_size_value)

sns.lineplot(data=df_psutil, x="timestamp", y="num_cameras", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
plt.show()

sns.lineplot(data=df_psutil, x=x_axis_label, y="CPU Percent", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("CPU load (all cores)")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/cpu/cpu_load.png")
plt.savefig("plots/cpu/cpu_load.pdf")
plt.show()

sns.lineplot(data=df_psutil, x=x_axis_label, y="GPU", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("GPU load")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/gpu/gpu_load.png")
plt.savefig("plots/gpu/gpu_load.pdf")
plt.show()

sns.lineplot(data=df_psutil, x=x_axis_label, y="GPU Watts", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
plt.ylabel("Power (watts)")
# plt.title("Average GPU power (watts)")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/gpu/gpu_power_watts.png")
plt.savefig("plots/gpu/gpu_power_watts.pdf")
plt.show()
```

```python

df_psutil = df_psutil.sort_values(by='timestamp')
df_psutil["timedelta"] = df_psutil["timestamp"].diff().apply(lambda x: min(5, x))
df_psutil["gpu_joules"] = df_psutil["timedelta"] * df_psutil["GPU Watts"]

sns.lineplot(data=df_psutil, x="num_cameras", y="gpu_joules", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
plt.show()

df_joules = df_psutil.groupby(["num_cameras", "sensor_size"])["gpu_joules"].sum().reset_index()
df_joules["joule_per_image"] = df_joules["gpu_joules"] / (df_joules["num_cameras"] * max(df_carla["num_frames"].unique()))

sns.lineplot(data=df_joules, x="num_cameras", y="gpu_joules", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
plt.show()

sns.lineplot(df_joules, x="num_cameras", y="joule_per_image", hue="sensor_size", palette=color_palette, hue_order=sensor_sizes)
plt.legend(title="Sensor Resolution")
# plt.title("GPU energy consumption per image (joules)")
plt.xlabel("Number of cameras")
plt.ylabel("Joules per image")
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("plots/gpu/gpu_joules_per_image.png")
plt.savefig("plots/gpu/gpu_joules_per_image.pdf")
plt.show()
```
